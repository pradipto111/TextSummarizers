```python
import requests
sample_text_url = 'https://raw.githubusercontent.com/pradipto111/TextSummarizer/main/sample_text.txt'
req = requests.get(sample_text_url)
text = req.text
print(text)
```

## IMPLEMENTATION FROM SCRATCH

```python
from nltk.tokenize import word_tokenize, sent_tokenize
import nltk, math, re
import numpy as np
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('stopwords')
from nltk.corpus import stopwords
nltk.download('punkt')
```

```python
sentences = sent_tokenize(text)
```

```python
t = text.lower()
t = t.replace(',','')
t = t.replace('’','')
filtered_sentences = sent_tokenize(t)
for i in range(len(filtered_sentences)):
  filtered_sentences[i] = filtered_sentences[i].replace('.','')
```

```python
luhn_score = []
for sent in filtered_sentences:
  words = word_tokenize(sent)
  begin = -1
  end = -1
  meaningful_word_count = 0
  for i in range(len(words)):
    word = words[i]
    if word not in stopwords.words('english'):
      meaningful_word_count += 1
      if begin != -1:
        begin = i
      end = i
  span = end-begin+1
  luhn_score.append((meaningful_word_count**2)/span)
luhn_score = np.asarray(luhn_score)
```

```python
summary_ratio = 0.2
summary_len = int(summary_ratio*len(sentences))
summary = ''
summary_idx = np.argsort(luhn_score)[-summary_len:]
summary_idx = summary_idx[::-1]
for i in summary_idx:
  summary += sentences[i]+'\n'
print(summary)
```

```python
from sumy.summarizers.luhn import LuhnSummarizer as Luhn
from sumy.nlp.tokenizers import Tokenizer as tokenizer
from sumy.parsers.plaintext import PlaintextParser
parser = PlaintextParser.from_string(text, tokenizer('english'))

summarizer = Luhn()
summary = summarizer(parser.document, sentences_count = 5)
text = ''
for sent in summary:
  text += str(sent)

print(text)
```
