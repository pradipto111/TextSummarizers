```python
import requests
sample_text_url = 'https://raw.githubusercontent.com/pradipto111/TextSummarizer/main/sample_text.txt'
req = requests.get(sample_text_url)
text = req.text
print(text)
```

## IMPLEMENTATION FROM SCRATCH

```python
from nltk.tokenize import word_tokenize, sent_tokenize
import nltk, math, re
import numpy as np
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('stopwords')
from nltk.corpus import stopwords
nltk.download('punkt')
```

```python
sentences = sent_tokenize(text)
```

```python
t = text.lower()
t = t.replace(',','')
t = t.replace('’','')
filtered_sentences = sent_tokenize(t)
for i in range(len(filtered_sentences)):
  filtered_sentences[i] = filtered_sentences[i].replace('.','')

t = t.replace('.','')
filtered_words = word_tokenize(t)
word_freq = {}
for word in filtered_words:
  if word not in stopwords.words('english'):
    if word not in word_freq:
      word_freq[word] = 1
    else:
      word_freq[word] += 1

max_freq = max(word_freq.values())

for word in word_freq:
  word_freq[word] /= max_freq

print(word_freq)
```

```python
sent_scores = []
for sent in filtered_sentences:
  sent_scores.append(0)
  for word in word_tokenize(sent):
    if word not in stopwords.words('english'):
      sent_scores[-1] += word_freq[word]

sent_scores = np.asarray(sent_scores)

summary_ratio = 0.2
summary_len = int(summary_ratio*len(sentences))
summary = ''
summary_idx = np.argsort(sent_scores)[-summary_len:]
summary_idx = summary_idx[::-1]
for i in summary_idx:
  print(sentences[i])
```
